```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = '/Users/pegaheizad/Desktop/Ghosh_lab/MyExpt_cells_filtered.csv'
data = pd.read_csv(csv_file)

# Convert to Excel file
excel_file = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files_2/MyExpt_cells_filtered.xlsx'
data.to_excel(excel_file, index=False)

# Read the Excel file
file_path = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files_2/MyExpt_cells_filtered.xlsx'
df = pd.read_excel(file_path)

# Create a copy of the dataframe
df_copy = df.copy()

# Add a new column "Normalized_Intensities"
df_copy['Normalized_Intensities'] = df_copy['Intensity_MeanIntensity_cd73_img'] / df_copy['Intensity_MeanIntensity_dapi_img']

# Save the modified dataframe to a new Excel file
output_file = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files_2/MyExpt_cells_filtered_normalized.xlsx'
df_copy.to_excel(output_file, index=False)

print(f"New Excel file with 'Normalized_Intensities' column has been saved as {output_file}.")

normalized_df = pd.read_excel('/Users/pegaheizad/Desktop/Ghosh_lab/data_files_2/MyExpt_cells_filtered_normalized.xlsx')
```

### TCP vs 527

```python
import pandas as pd
import matplotlib.pyplot as plt

# Separate data based on growing surface
tcp_data = normalized_df[normalized_df['Metadata_FileLocation'].str.contains('TCP', na=False)]['Normalized_Intensities']
soft_data = normalized_df[normalized_df['Metadata_FileLocation'].str.contains('soft', na=False)]['Normalized_Intensities']

# Create the overlayed histograms
plt.figure(figsize=(8, 6))
plt.xlim(left=0, right=10)
plt.hist(tcp_data, bins=30, alpha=0.5, label='TCP (Hard Surface)', color='blue')
plt.hist(soft_data, bins=30, alpha=0.5, label='Soft Surface', color='orange')

# Add titles and labels
plt.title('Histogram of Normalized Mean Intensity Across Growing Surfaces')
plt.xlabel('Normalized Mean Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the excel file
file_path = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files/cd73_data.xlsx'
df = pd.read_excel(file_path)

# Separate data based on P3 & growing surface
p3_data = df['Intensity_MeanIntensity_cells']

# Calculate the mean, median, and standard deviation
mean_p3_data = p3_data.mean()
median_p3_data = p3_data.median()
std_p3_data = p3_data.std()

# Print the results
print(f"Mean Intensity: {mean_p3_data}")
print(f"Median Intensity: {median_p3_data}")
print(f"Standard Deviation of Intensity: {std_p3_data}")
```

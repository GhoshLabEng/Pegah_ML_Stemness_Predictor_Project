```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the excel file
file_path = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files/cd105_selected_columns.xlsx'
df = pd.read_excel(file_path)

# Separate data based on growing surface
tcp_data = df[df['Metadata_GrowingSurface'] == 'TCP']['Intensity_MeanIntensity_cells']
soft_data = df[df['Metadata_GrowingSurface'] == 'soft']['Intensity_MeanIntensity_cells']

# Create the overlayed histograms
plt.figure(figsize=(8, 6))
plt.hist(tcp_data, bins=30, alpha=0.5, label='TCP (Hard Surface)', color='blue')
plt.hist(soft_data, bins=30, alpha=0.5, label='Soft Surface', color='orange')

# Add titles and labels
plt.title('Histogram of Mean Intensity Across Growing Surfaces')
plt.xlabel('Mean Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the excel file
file_path = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files/cd105_selected_columns.xlsx'
df = pd.read_excel(file_path)

# Separate data based on growing surface
p3TCP_data = df[(df['Metadata_Passage'] == 3) & (df['Metadata_GrowingSurface'] == 'TCP')]['Intensity_MeanIntensity_cells']
p3soft_data = df[(df['Metadata_Passage'] == 3) & (df['Metadata_GrowingSurface'] == 'soft')]['Intensity_MeanIntensity_cells']
p5TCP_data = df[(df['Metadata_Passage'] == 5) & (df['Metadata_GrowingSurface'] == 'TCP')]['Intensity_MeanIntensity_cells']
p5soft_data = df[(df['Metadata_Passage'] == 5) & (df['Metadata_GrowingSurface'] == 'soft')]['Intensity_MeanIntensity_cells']

# Create the overlayed histograms
plt.figure(figsize=(8, 6))
plt.hist(p3TCP_data, bins=30, alpha=0.5, label='P3 TCP', color='blue')
plt.hist(p3soft_data, bins=30, alpha=0.5, label='P3 soft', color='orange')
plt.hist(p5TCP_data, bins=30, alpha=0.5, label='P5 TCP', color='pink')
plt.hist(p5soft_data, bins=30, alpha=0.5, label='P5 soft', color='green')

# Add titles and labels
plt.title('Histogram of Mean Intensity Across Passages & Growing Surfaces')
plt.xlabel('Mean Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

### P3 soft 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the excel file
file_path = '/Users/pegaheizad/Desktop/Ghosh_lab/data_files/cd105_selected_columns.xlsx'
df = pd.read_excel(file_path)

# Separate data based on P3 & growing surface
p3soft_data = df[(df['Metadata_Passage'] == 3) & (df['Metadata_GrowingSurface'] == 'soft')]['Intensity_MeanIntensity_cells']

# Calculate the mean, median, and standard deviation
mean_p3soft_data = p3soft_data.mean()
median_p3soft_data = p3soft_data.median()
std_p3soft_data = p3soft_data.std()

# Print the results
print(f"Mean Intensity: {mean_p3soft_data}")
print(f"Median Intensity: {median_p3soft_data}")
print(f"Standard Deviation of Intensity: {std_p3soft_data}")
```

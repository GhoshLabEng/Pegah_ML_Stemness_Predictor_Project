### TCP vs 527

```python
import pandas as pd
import matplotlib.pyplot as plt

# Separate data based on growing surface
tcp_data = df[df['Metadata_FileLocation'].str.contains('TCP', na=False)]['Intensity_MeanIntensity_cd73_img']
soft_data = df[df['Metadata_FileLocation'].str.contains('soft', na=False)]['Intensity_MeanIntensity_cd73_img']

# Create the overlayed histograms
plt.figure(figsize=(8, 6))
plt.xlim(left=0, right=0.2)
plt.hist(tcp_data, bins=30, alpha=0.5, label='TCP (Hard Surface)', color='blue')
plt.hist(soft_data, bins=30, alpha=0.5, label='Soft Surface', color='orange')

# Add titles and labels
plt.title('Histogram of Mean Intensity Across Growing Surfaces')
plt.xlabel('Mean Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Read the original CSV file
file_path = '/Users/pegaheizad/Desktop/Enzyme-inhibitors-grades.csv'
df = pd.read_csv(file_path)

# Create an empty DataFrame to hold the modified data
new_df = pd.DataFrame(columns=df.columns)

# Iterate through the original DataFrame and append rows and empty rows
for i in range(len(df)):
    new_df = pd.concat([new_df, pd.DataFrame([df.iloc[i]], columns=df.columns)], ignore_index=True)  # Add original row
    new_df = pd.concat([new_df, pd.DataFrame([[None]*len(df.columns)], columns=df.columns)], ignore_index=True)  # Add empty row

# Save the new DataFrame to a CSV file
new_file_path = '/Users/pegaheizad/Desktop/Enzyme-inhibitors-grades-with-empty-rows.csv'
new_df.to_csv(new_file_path, index=False)

print(f"New CSV saved to {new_file_path}")


```
